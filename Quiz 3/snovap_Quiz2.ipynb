````
plt.figure(figsize=(6, 4))

# Creating histogram
plt.hist(titanic['Age'].dropna(), bins=20, color='skyblue', edgecolor='black')

#labels
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')

plt.ylim(0, 100)

plt.show()
````

En aquella época, la sociedad tenía una marcada desigualdad de género, lo que explica que la mayoría de las personas a bordo del Titanic fueran hombres. Sin embargo, como se retrata en la película, durante la evacuación se priorizó a mujeres y niños para abordar los botes salvavidas, siguiendo el principio de "primero las mujeres y los niños". Esto resultó en una mayor proporción de mujeres entre los sobrevivientes.

````
from statsmodels.graphics.mosaicplot import mosaic

# Filter relevant columns
filtered_data = titanic[['Sex', 'Survived']]

# Calculate proportions
grouped_data = filtered_data.groupby(['Sex', 'Survived']).size().unstack()

# Change the order of columns
grouped_data = grouped_data[[1, 0]]

# Rename columns
grouped_data.columns = ['Survived', 'Died']

# Change gender labels
grouped_data.index = ['Female' if index == 'female' else 'Male' for index in grouped_data.index]

# Create a mosaic plot
mosaic(grouped_data.stack(), axes_label=False, title="Proportions of survival by sex")

# Show the plot
plt.show()
````

Es comprensible que, en medio de la tragedia, las personas con mayor autoridad o poder en el Titanic hayan tenido más acceso a las medidas de supervivencia. A partir de este gráfico y el anterior, podemos deducir que la gran mayoría de los hombres a bordo pertenecían a la clase trabajadora y poseían boletos de tercera clase. Esto también explica por qué la proporción de mujeres fallecidas fue menor, ya que muchas de las mujeres que viajaban eran de clase alta. En esa época, las mujeres no realizaban trabajos manuales, por lo que, si no viajaban como acompañantes con boletos de primera clase, es probable que no abordaran el barco.

````

# Filter relevant columns
filtered_data = titanic[['Pclass', 'Survived']]

# Calculate proportions
grouped_data = filtered_data.groupby(['Pclass', 'Survived']).size().unstack()

# Change the order of columns
grouped_data = grouped_data[[1, 0]]

# Rename columns
grouped_data.columns = ['Survived', 'Died']

# Create a mosaic plot
mosaic(grouped_data.stack(), axes_label=False, title='Survival rate by class ticket')

# Show the plot
plt.show()
````

Las tasas de supervivencia en cada puerto son bastante uniformes. En Southampton, donde abordó la mayor cantidad de pasajeros, probablemente se concentró la mayor proporción de trabajadores con boletos de tercera clase. En los siguientes puertos, es posible que hubiera una mayor presencia de viajeros y turistas, es decir, pasajeros con boletos de primera clase, quienes, según el gráfico anterior, tenían una mayor probabilidad de sobrevivir.

````
filtered_data = titanic[['Embarked', 'Survived']]

# Calculate proportions
grouped_data = filtered_data.groupby(['Embarked', 'Survived']).size().unstack()

# Change the order of columns
grouped_data = grouped_data[[1, 0]]

# Rename columns
grouped_data.columns = ['Survived', 'Died']

# Create a mosaic plot
mosaic(grouped_data.stack(), axes_label=False, title="Survival rates by port of embarcation")

# Show the plot
plt.show()
````
